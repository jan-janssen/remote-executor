```python
from executorlib_cache import FileExecutor
```

```python
from executorlib_cache.shared import execute_in_subprocess
```

```python
from pysqa import QueueAdapter
```

```python
def my_funct(a, b):
    return a+b
```

```python
def execute_with_pysqa(command, task_dependent_lst=[]):
    qa = QueueAdapter(directory="resources/queues")
    return qa.submit_job(command=" ".join(command), working_directory=".")
```

```python
with FileExecutor(execute_function=execute_with_pysqa) as exe:
    fs1 = exe.submit(my_funct, 1, b=2)
    print(fs1.done(), fs1.result(), fs1.done())
    # fs2 = exe.submit(my_funct, 1, b=fs1)
    # print(fs2.done(), fs2.result(), fs2.done())
```
